```python
df_salary = pd.read_csv("Salaries.csv")
df_salary.head()
```

```python
print(f"Các loại nghề nghiệp hiện có : {df_salary.JobTitle.unique()}")
print(f"Tổng số các nghề là: {len(df_salary.JobTitle.unique())}")
```

```python
df_salary.BasePay = pd.to_numeric(df_salary.BasePay, errors='coerce')
arrYearGroupBy = df_salary[["Year","BasePay"]].groupby("Year",as_index = False).mean()
print(f"Lương trung bình theo năm là")
display(arrYearGroupBy)
```

```python
print(f"Shape 0 của df: {len(df_salary)}")
print(f"{df_salary.columns}")
Condition1 = df_salary[(df_salary.EmployeeName == "Not provided") | (df_salary.EmployeeName == "Not Provided")]
arrDropName = df_salary.drop(Condition1.index)
display(arrDropName)
```
