```python
import pandas as pd
from sklearn.datasets import fetch_openml
df, y = fetch_openml('titanic', version=1, as_frame=True,return_X_y=True)
print(df.shape)
df.head(3)
```

```python
?fetch_openml
```

```python
print(df.info())
print(df.isnull().sum())
```

```python
print(f"Kích thước dữ liệu: {df.shape}")
df.drop(['cabin','boat', 'body', 'home.dest'],axis=1, inplace = True)

print(f"Kích thước sau khi loại bỏ dữ liệu dư thừa: {df.shape}")
print(df.info)
```

```python
from sklearn.impute import SimpleImputer
print(f'Số lượng giá trị trống: {df.age.isnull().sum()}')

imp = SimpleImputer(strategy= 'mean')
df['age'] = imp.fit_transform(df[['age']])
print(f"Số lượng giá trị trống sau khi biến đổi: {df.age.isnull().sum()}")
df.age.head(10)
```

```python
df['family'] = df['sibsp'] + df['parch']
df.loc[df['family'] > 0, 'travelled_alone'] = 0
df.loc[df['family'] == 0, 'travelled_alone'] = 1
df['travelled_alone'].value_counts().plot(title='Pasenger travelled alone?',kind='bar')
```

```python
from sklearn.preprocessing import OneHotEncoder
df['sex'] = OneHotEncoder().fit_transform(df[['sex']]).toarray()[:,1]
df.sex.head(10)
```

```python
cat_cols = df.select_dtypes(include=['category']).columns
print(f'Categorical columns: {cat_cols}')
 
# Loop through each categorical column 
for col in cat_cols:
    # Impute missing values
    fill_value = df[col].mode()[0] 
    df[col].fillna(fill_value, inplace=True)
     
    # create a list of labels to be encoded in the column
    append_to = list(df[col].unique())
 
    # These labels will be use as column headers
    print(append_to)
 
    # Apply OneHotEncoder()
    df[append_to] = OneHotEncoder().fit_transform(df[[col]]).toarray()
 
    # Drop non-encoded column
    df.drop(col, axis=1, inplace=True)
 
    # Drop redundant data
    df.drop(append_to[0], axis=1, inplace=True)

df.head()
```

```python
from sklearn.preprocessing import MinMaxScaler
num_cols = df.select_dtypes(include=['int64', 'float64', 'int32']).columns
print(num_cols)
 
# Impute missing values
for col in num_cols:
    fill_value = df[col].mean()
    df[col].fillna(fill_value, inplace= True)

minmax = MinMaxScaler()
df[num_cols] = minmax.fit_transform(df[num_cols])
df[num_cols]
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt
```

# Bài 1 & 2: sử dụng mô hình linear logistic và pipeline đối với dữ liệu sau tiền xử lý

```python
df.head()
```

```python
df.drop(['name','ticket'], axis=1, inplace=True)
```

```python
df.info()
```

```python
X_train, X_test, y_train, y_test = train_test_split(df, y, test_size = 0.2, random_state = 0)
classifier = LogisticRegression(random_state = 6, fit_intercept = True)
classifier.fit(X_train, y_train)

y_pred = classifier.predict(X_test)
PredictSurvive = confusion_matrix(y_test, y_pred)
PredictSurvive = pd.DataFrame(PredictSurvive)
```

```python
PredictSurvive
```

```python
# qd = pd.DataFrame(qd)
#display("Confusion Matrix : \n", qd)

ax = sns.heatmap(PredictSurvive, annot=True, cmap="BuPu", fmt='g')
ax.set_title("Confusion Matrix for data");
ax.set_xlabel("Prediction for survival")
ax.set_ylabel('Actual survival');
ax.xaxis.set_ticklabels(["No","Yes"])
ax.yaxis.set_ticklabels(["No","Yes"])
plt.show()
```

# Bài 3: Thực hành các hoạt động Tiền xử lý dữ liệu với dữ liệu datasets.load_iris()

```python
from sklearn.datasets import load_iris
```

```python
x,y = load_iris(return_X_y=True, as_frame=True)
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.4, random_state=5)
print(X_train.shape)
print(y_train.shape)
print(X_test.shape)
print(y_test.shape)
x.head()
```

```python
y.unique()
```

```python
ClassIris = load_iris().target_names
print(ClassIris)
```

```python
x.info()
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn import metrics

# experimenting with different n values
k_range = list(range(1,26))
scores = []
for k in k_range:
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(X_train, y_train)
    y_pred = knn.predict(X_test)
    scores.append(metrics.accuracy_score(y_test, y_pred))
    
plt.plot(k_range, scores)
plt.xlabel('Value of k for KNN')
plt.ylabel('Accuracy Score')
plt.title('Accuracy Scores for Values of k of k-Nearest-Neighbors')
plt.show()
```
